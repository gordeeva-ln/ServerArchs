```python
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(3, 3, figsize=(16, 8));

axes[0][0].set_ylabel('Sort')
axes[1][0].set_ylabel('Client on server')
axes[2][0].set_ylabel('Client')

axes[2][0].set_xlabel('N')
axes[2][1].set_xlabel('M')
axes[2][2].set_xlabel('Delta')

crit = ['N', 'M', 'D']

for i in range(3):
    for j in range(3):
        for k in range(3):
            array = []
            with open("res_{}_{}_{}.txt".format(k + 1, crit[j], i + 1)) as f:
                first = True
                for score in f:
                    if first:
                        first = False
                        x, minc, maxc, step, n, m, d = map(lambda cr: float(cr), score.split())
                    else:    
                        array.append(float(score))
            axes[i][j].plot([minc + step * ic for ic in range(int((maxc - minc) / step))], array)
fig.legend(labels=["Server1", "Server2", "Server3"],
           loc="center right",
           borderaxespad=0.1,
           title="Legend Title"
           )

plt.show()
```
